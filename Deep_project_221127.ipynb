```python
import numpy as np
import pandas as pd
import tensorflow as tf

import autokeras as ak
from sklearn.model_selection import train_test_split
```

```python
base_path = '/content/drive/MyDrive/Colab Notebooks/MachineLearning/data/Construction_machinery/'


train = pd.read_csv(base_path + 'train.csv')
test = pd.read_csv(base_path + 'test.csv')
submission = pd.read_csv(base_path + 'sample_submission.csv')
```

```python
target = ['ANONYMOUS_1', 'YEAR', 'ANONYMOUS_2', 'AG',
           'CO', 'CR', 'CU', 'FE', 'H2O', 'MN', 'MO', 'NI', 'PQINDEX', 'TI', 'V',
           'V40', 'ZN','Y_LABEL'] # test 내부 컬럼이 모두 사용됬습니다.

           
comp4 = train.loc[:,target]
train4, test4 = train_test_split(comp4, test_size=0.15, random_state=21)
train4.to_csv(f'./comp_train.csv')
test4.to_csv(f'./comp_test.csv')

target = ['ANONYMOUS_1', 'YEAR', 'ANONYMOUS_2', 'AG',
           'CO', 'CR', 'CU', 'FE', 'H2O', 'MN', 'MO', 'NI', 'PQINDEX', 'TI', 'V',
           'V40', 'ZN']


tp = test.loc[:,['ID']]
test = test.loc[:,target]
test.to_csv('./test_target.csv')

###############################################

clf = ak.StructuredDataClassifier(
    max_trials=10,  # 10회 생성
    overwrite=True,
)
# 몇가지 옵션요소가 있습니다.
# https://autokeras.com/

clf.fit(
    f'./comp_train.csv', # train path
    "Y_LABEL", # label
    epochs=500,
)

predicted_y = clf.predict(f'./comp_test.csv') # valid path
# print(clf.evaluate(f'./comp{x}_test.csv', "Y_LABEL"))
model = clf.export_model()
pred = model.predict(pd.read_csv('./test_target.csv')) # test path
```

```python
aa = []
for i in pred:
    if i[0] > 0.5: # huddle
        aa.append(1)
    else:
        aa.append(0)

tp['Y_LABEL'] = aa

# submission 형식을 맞춰주세요
# tp = tp.set_index('ID')

tp.to_csv('./comp_ans.csv')
pd.read_csv('./comp_ans.csv').Y_LABEL.value_counts()
```
